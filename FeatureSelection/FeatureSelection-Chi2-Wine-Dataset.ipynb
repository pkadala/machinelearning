```python
import pandas as pd
import numpy as np

from sklearn import datasets
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2, f_classif, f_regression
```

```python
wine_dataset = datasets.load_wine()

wine_df = pd.DataFrame(wine_dataset.data, columns=wine_dataset.feature_names)
wine_df.describe()
```

```python
# Create and fit selector
selector = SelectKBest(chi2)
selector.fit(wine_df, wine_dataset.target)
# Get columns to keep and create new dataframe with those only
cols = selector.get_support(indices=True)
wine_df_new = wine_df.iloc[:,cols]

print('Original number of features:', wine_df.shape)
print(wine_df.columns.values)
print('\nReduced number of features:', wine_df_new.shape)
print(wine_df_new.columns.values)


```

```python
f_values, p_values = chi2(wine_df, wine_dataset.target)
reg_summary = pd.DataFrame(wine_df.columns.values, columns=['Features'])
reg_summary ['f-values'] = f_values.round(3)
reg_summary ['p-values'] = p_values.round(3)
reg_summary
```
